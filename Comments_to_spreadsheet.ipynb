```python
#creating header to avoid errors upon requests of webpages
header = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) '
      'AppleWebKit/537.11 (KHTML, like Gecko) '
      'Chrome/23.0.1271.64 Safari/537.11',
      'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
      'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
      'Accept-Encoding': 'none',
      'Accept-Language': 'en-US,en;q=0.8',
      'Connection': 'keep-alive'}
```

```python
#function to get pdf from page
def pdf_text(url, header_accept):

  response = requests.get(url)

  pdf_reader = fitz.open(stream=response.content, filetype="pdf")

  raw_pdf = ''
  for page_num in range(pdf_reader.page_count):
    content = pdf_reader.load_page(page_num)
    if content:
      raw_pdf += content.get_text()
    raw_pdf = raw_pdf.lstrip()

  return raw_pdf
```

```python
def add_comment_text(df):
  df['download_link'] = None
  df['comment_link'] = None
  df['comment'] = None

  count = 0
  for index, row in df.iterrows():

    #getting data
    comment_id = row['id']
    comment_url = f'https://api.regulations.gov/v4/comments/{comment_id}?include=attachments&api_key={reg_api_key}'
    r = http.request('GET', comment_url)
    comment_data = json.loads(r.data.decode('utf-8'))

    df.loc[index, 'comment_link'] = f'https://www.regulations.gov/comment/{comment_id}'
    if 'included' not in comment_data:
      df.loc[index, 'comment'] = comment_data['data']['attributes']['comment']
      df.loc[index, 'download_link'] = None
    else:
      try:

        download_link = comment_data['included'][0]['attributes']['fileFormats'][0]['fileUrl']

        comment = pdf_text(download_link, header)

        df.loc[index, 'comment'] = comment
        df.loc[index, 'download_link'] = download_link
        count +=1
        print('*'*100)
        print(f'Documents added: {count}')
      except Exception as e:
        print("An error occured:", e)
        print("Continuing")





```
